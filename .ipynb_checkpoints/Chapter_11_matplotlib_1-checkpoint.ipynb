```python
import matplotlib.pyplot as plt
```

```python
x = [2,5,8]
y = [6,2,18]
y2 = [8,19,4]
plt.plot(x,y,label='First',color='r',linewidth=3)
plt.plot(x,y2,label='Second',color='c',linewidth=3)
plt.xlabel('X (time)')
plt.ylabel('y ($)')
plt.title('Chart\n nemodar')
plt.legend()
plt.show()
```

```python
age = [22,34,38,55,16,18,29,45,46,72,35,12]
bins = [0,20,40,60,80]
plt.hist(age,bins,histtype='bar',rwidth=0.4)
plt.show()
```

```python
x = [2,5,8,11,14,15,18,22]
y = [6,2,18,13,16,10,19,25]
plt.scatter(x,y,label='test',color='g',s=100,marker='x')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter')
plt.show()
```

```python
major = ['eng','medic','art','lower']
stu = [120,80,42,53]
plt.pie(stu,labels=major,colors=['c','r','g','m'],explode=(.1,0.1,0,0),autopct='%1.1f%%')
plt.title('University')
plt.show()
```
